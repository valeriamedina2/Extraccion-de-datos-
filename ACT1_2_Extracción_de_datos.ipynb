```python
df= pd.read_excel("Datos de Facturación(1).xlsx")
df
```

CVE_CLPV (valores 10000 a 2000)

```python
filtro1 =df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1
```

```python
filtro1.to_csv("filtrovalores2.csv")
#descargar archivo filtrado en csv
from google.colab import files   
files.download("filtrovalores2.csv")
```

CVE_VEND(todas las claves excepto "5" y "4") 

```python
filtro2=df[~(df["CVE_VEND"] == 5) & ~(df["CVE_VEND"] == "4")]
filtro2
```

```python
filtro2.to_csv("filtronot2.csv")
#descargar archivo filtrado en csv
from google.colab import files   
files.download("filtronot2.csv")
```

FECHA_ENT (Fechas de 28/02/2022)

```python
filtro3=df[df["FECHA_ENT"] == "28-02-2022  0:00:00"]
filtro3
```

```python
filtro3.to_csv("filtrofechas.csv")
#descargar archivo filtrado en csv
from google.colab import files   
files.download("filtrofechas.csv")
```

CAN_TOT (Cantidades menores a 5951.7) o STATUS ("E")

```python
filtro4=df[(df["CAN_TOT"] > 5951.7) | (df["STATUS"] == "E")]
filtro4
```

```python
filtro4.to_csv("filtroo2.csv")
#descargar archivo filtrado en csv
from google.colab import files   
files.download("filtroo2.csv")
```
